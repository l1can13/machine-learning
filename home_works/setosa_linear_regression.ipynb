```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
setosa_data = iris.data[iris.target == 0]
```

```python
X = setosa_data[:, 1]  # sepal_width
y = setosa_data[:, 3]  # petal_width
```

```python
# Функция для линейной регрессии
def linear_regression(X, y):
    mean_x = np.mean(X)
    mean_y = np.mean(y)
    m = np.sum((X - mean_x) * (y - mean_y)) / np.sum((X - mean_x)**2)
    b = mean_y - m * mean_x
    return m, b
```

```python
# Получаем коэффициенты наклона и y-пересечения
slope, intercept = linear_regression(X, y)
```

```python
# Выводим результаты
print(f'Наклон (slope): {slope}')
print(f'y-пересечение (intercept): {intercept}')
```

```python
# Построение линейной регрессии
regression_line = slope * X + intercept
```

```python
# Построение графика
plt.scatter(X, y, label='Данные')
plt.plot(X, regression_line, color='red', label='Линейная регрессия')
plt.xlabel('Sepal Width')
plt.ylabel('Petal Width')
plt.title('Линейная регрессия для setosa')
plt.legend()
plt.show()
```
